# アイスクリームの支出と気温のグラフ

*Python3*を使用して、**グラフ**を<font color='red'>作成</font>します。

## ライブラリの読み込み

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## データ入力

```python
icecream=pd.DataFrame([[1,464,10.6],[2,397,12.2],[3,493,14.9],[4,617,20.3],[5,890,25.2],[6,883,26.3],[7,1292,29.7],[8,1387,31.6],[9,843,27.7],[10,621,22.6],[11,459,15.5],[12,561,13.8]],columns=['Month','Monthly ice cream expenditure','Mean temperature'])
```

## DF.plot で折れ線グラフを描画

```python
icecream['Monthly ice cream expenditure'].plot()
plt.show()
```

## 表タイトルや軸ラベルを付ける

```python
icecream['Monthly ice cream expenditure'].plot()
plt.title("Monthly ice cream expenditure per household")
plt.xlabel('Month')
plt.ylabel('Ice cream expenditure(Jpapanese yen)')
plt.show()
```

## plt.plot(DF)でグラフを描画

```python
plt.plot(icecream[['Monthly ice cream expenditure']])
plt.title("Monthly ice cream expenditure per household")
plt.xlabel('Month')
plt.ylabel('Ice cream expenditure(Jpapanese yen)')
plt.show()
```

## 2つのデータを描き、凡例を付ける

```python
icecream[['Monthly ice cream expenditure','Mean temperature']].plot(legend=True)
plt.title("Monthly ice cream expenditure per household")
plt.xlabel('Month')
plt.ylabel('Ice cream expenditure(Jpapanese yen)')
plt.show()
```

## 棒グラフで2つデータを描画

```python
icecream[['Monthly ice cream expenditure','Mean temperature']].plot(kind='bar')
plt.title("Monthly ice cream expenditure per household")
plt.xlabel('Month')
plt.ylabel('Ice cream expenditure(Jpapanese yen)')
plt.show()
```

## 散布図で2つのデータを描画

```python
icecream.plot.scatter(y='Monthly ice cream expenditure',x='Mean temperature')
plt.title("Monthly ice cream expenditure per household")
plt.xlabel('Mean temperature')
plt.ylabel('Ice cream expenditure(Jpapanese yen)')
plt.show()
```

## バブルチャートで2つのデータを描画

```python
icecream.plot.scatter(s=icecream['Monthly ice cream expenditure']*0.5,x='Month',y='Mean temperature')
plt.title("Monthly ice cream expenditure per household")
plt.xlabel('Month')
plt.ylabel('Mean temperature')
plt.show()
```

## 月-気温-アイスクリーム支出を色で描く

```python
icecream.plot.scatter(x='Month',y='Mean temperature',s=81,c=icecream['Monthly ice cream expenditure']*0.5,cmap='bwr',edgecolor='k')
plt.title("Monthly ice cream expenditure per household")
plt.xlabel('Month')
plt.ylabel('Mean temperature')
plt.show()
```
